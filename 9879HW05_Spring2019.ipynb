## The Kyle (1985) model

### 1. (14 points)

Suppose that the monopolistic trader in the Kyle (1985) model is not told the true value $\tilde S$ but is instead given a signal $x=\tilde S+\epsilon$ with $\epsilon \sim N(0,\sigma_\epsilon^2)$, where $\tilde S$ and $\epsilon$ are assumed independent.

(a) Compute the conditional expectation $\mathbb{E}[\tilde S|x]$.

---

$x$ is normally distributed with mean $\mu$ and variance $\sigma^2 + \sigma_e^2$, the conditional expectation $\mathbb{E}[\tilde S|x]$ is given by 
$$
\begin{align*}
\mathbb{E}[\tilde S|x] &= \mathbb{E}[\tilde S] + \frac{Cov(\tilde S, x)}{Var(x)}(x-\mathbb{E}[x])\\
\end{align*}
$$
where $\mathbb{E}[\tilde S] = \mathbb{E}[x]=\mu$, $Var(x) = \sigma^2 + \sigma_e^2$ and
$$
\begin{align*}
Cov(\tilde S, x) =& Cov(\tilde S, \tilde S + \epsilon)\\
=& Cov(\tilde S, \tilde S) + Cov(\tilde S,\epsilon)\\
=& Var(\tilde S) + 0\\
=& \sigma^2\\
\end{align*}
$$

Therefore,
$$
\begin{align*}
\mathbb{E}[\tilde S|x] =& \mu + \frac{\sigma^2}{\sigma^2 + \sigma_e^2}(x-\mu)\\
=& \mu + \phi (x-\mu)\\
\end{align*}
$$

---

(b) If $\mathcal I$ assumes that the market maker $\mathcal M$ generates a price function of the form $p=\mu +\lambda\,y$, show that the informed trader's optimal (profit maximizing) demand is given by

$$
y_i=\frac{\phi}{2\,\lambda}\,\left(x-\mu\right)
$$

where

$$
\phi = \frac{\sigma^2}{\sigma^2+\sigma_\epsilon^2}.
$$

---


Knowing that the total demand $y$ is given by the sum of the demand from informed ($y_i$) and uninformed traders ($y_u$) and that $y_u$ has mean $0$ and is independent of $\tilde S$, the expected profit of $\mathcal I$ is:
$$
\begin{align*}
\pi =& \mathbb{E}[y_i(\tilde S - p)|x]\\
=& \mathbb{E}[y_i(\tilde S - \mu - \lambda(y_i+y_u))|x]\\
=& y_i(\mu + \phi(x-\mu) - \mu - \lambda y_i)\\
=& y_i(\phi(x-\mu) - \lambda y_i)\\
\end{align*}
$$

Maximizing the expected profit $\pi$, the first-order condition is:
$$
\begin{align*}
&\phi(x-\mu) - 2\lambda y_i = 0\\
\implies &y_i = \frac{\phi}{2\lambda}(x-\mu)\\
\end{align*}
$$

(c) Following the derivation from the lecture slides, use the result of (b) to deduce equation (11) in the lecture slides:

$$
\lambda=\frac{\sigma}{2\,\sigma_u}\,\sqrt{\phi}.
$$

---
$\mathcal M$ thinks that the demand function of $\mathcal I$  is $y_i = a+bx$ and sets the pricing to $p = \mu + \lambda y = \mathbb{E}[\tilde S | y]$. 
To calculate this conditional expectation as in the part a), we need to find the variance of $y$ and the covariance of $\tilde S$ and $y$:

$$
\begin{align*}
Var(y_i) =& Var\left(\frac{\phi}{2\lambda}(x-\mu)\right)\\
=& \frac{\phi^2}{4\lambda^2}(\sigma^2 + \sigma_e^2)\\
=& \frac{\phi\sigma^2}{4\lambda^2}\\
\end{align*}
$$

$$
\begin{align*}
Cov(\tilde S, y_i) =& \frac{\phi\sigma^2}{2\lambda}\\
\end{align*}
$$

Then the expectation $\mathbb{E}[\tilde S | y]$ is:
$$
\begin{align*}
\mathbb{E} [\tilde S|y] =& \mathbb{E} [\tilde S]+\frac{Cov(\tilde S,y)}{Var(y)}\left(y-\mathbb{E} [y]\right)\\
=& \mu + \frac{\frac{\phi\sigma^2}{2\lambda}}{\sigma_u^2 + \frac{\phi\sigma^2}{4\lambda^2}}y\\
\end{align*}
$$

Solve for the Kyle lambda:
$$
\begin{align*}
&\lambda = \frac{\frac{\phi\sigma^2}{2\lambda}}{\sigma_u^2 + \frac{\phi\sigma^2}{4\lambda^2}}\\
\implies & \lambda = \frac{2\lambda\phi\sigma^2}{4\lambda^2\sigma_u^2 + \phi\sigma^2}\\
\implies & 4\lambda^2\sigma_u^2 + \phi\sigma^2 = 2\phi\sigma^2\\
\implies & \lambda^2  = \frac{\phi\sigma^2}{4\sigma_u^2}\\
\implies & \lambda  = \frac{\sigma}{2\sigma_u}\sqrt\phi\\
\end{align*}
$$

### 2. (2 points) 

In the Kyle model, if the characteristic uninformed demand $\sigma_u$ increases by some factor $\theta$, what happens to the optimal informed trader demand $y_i$?

---
The optimal informed trader demand is 
$$
y_i = \frac{\sigma_u}\sigma(\tilde S - \mu).
$$
